Desenvolver um programa para verificar a nota do aluno em uma prova com 10 questões, o programa deve perguntar ao aluno a resposta de cada questão e ao final comparar com o gabarito da prova e assim calcular o total de acertos e a nota (atribuir 1 ponto por resposta certa). Após cada aluno utilizar o sistema deve ser feita uma pergunta se outro aluno vai utilizar o sistema. Após todos os alunos terem respondido informar: <br><br>
a) Maior e Menor Acerto; <br>
b) Total de Alunos que utilizaram o sistema; <br>
c) A Média das Notas da Turma. <br><br>
Gabarito da Prova: <br>
<br>
01 - A <br>
02 - B <br>
03 - C <br>
04 - D <br>
05 - E <br>
06 - E <br>
07 - D <br>
08 - C <br>
09 - B <br>
10 - A <br><br>

```python
gabarito = ['A', 'B', 'C', 'D', 'E', 'E', 'D', 'C', 'B', 'A'] # lista com o gabarito
lista_respostas = [] # lista que armazenará as respostas do aluno
nota = 0 # contador para definir nota do aluno
lista_notas = [] # lista que armazenará a nota de cada aluno
total_alunos = 0 # contador para saber quantos alunos utilizaram o sistema

cont = 0 # contador inicial das 10 respostas dos alunos

while cont == 0: # enquanto o contador inicial for igual a 0:
    i = 1 # definir índice como 1
    while i <= 10: # enquanto índice for menor ou igual a 10
        resposta = input(f'Resposta da {i}ª questão: ').upper() # pedir para aluno inserir resposta à pergunta
        lista_respostas.append(resposta) # adicionar a resposta na lista_respostas
        i += 1 # adicionar 1 ao índice (para que o loop peça a resposta da próxima pergunta até chegar na 10ª)

    for i, resp in enumerate(lista_respostas): # para cada item da lista de respostas inseridas pelo aluno:
        if resp == gabarito[i]: # se a resposta for igual ao gabarito
            nota += 1 # acrescentar 1 ponto à nota

    lista_notas.append(nota) # adicionar nota à lista_notas
    total_alunos += 1 # adicionar 1 ao contador de quantos alunos utilizaram o sistema

    cont2 = 0 # contador final para saber se o sistema continuará sendo usado
    while cont2 == 0: # enquanto o contador final for igual a zero:
        continuar = input('Outro aluno utilizará este sistema? Digite "S" (sim) ou "N" (não): ').upper() # pergunta se outro aluno utilizará o sistema
        if continuar == 'S': # se a resposta for sim:
            lista_respostas.clear()  # limpar lista_respostas para armazenar as respostas do próximo aluno
            nota = 0 # zerar nota
            cont = 0 # zerar contador inicial
            cont2 += 1 # acrescentar 1 ao contador final (para sair deste loop e voltar ao primeiro que pede as notas ao usuário)
        elif continuar == 'N': # se a resposta for não, calcular e imprimir na tela:
            for i, acerto in enumerate(lista_notas): # para cada item da lista de notas:
                acerto == max(lista_notas) # encontrar a maior nota
                print(f'\nA maior nota foi {max(lista_notas)}.') # imprimir na tela a maior nota
                acerto == min(lista_notas) # encontrar a menor nota
                print(f'A menor nota foi {min(lista_notas)}.') # imprimir na tela a menor nota
                print(f'{total_alunos} alunos utilizaram este sistema.') # imprimir na tela o total de alunos que utilizaram o sistema          
                print(f'A média de notas desta turma foi de {sum(lista_notas) / len(lista_notas):.2f}.') # imprimir na tela a média dos alunos da turma
                cont += 1 # acrescentar 1 ao contador inicial (para que não volte para o primeiro loop que pede mais notas, já que o usuário informou "não")
                cont2 += 1 # acrescentar 1 ao contador final (para sair deste loop, já que o usuário informou "não")
                break # encerrar o loop
        else: # se a resposta não for sim nem não:
            print('Resposta inválida, tente novamente.') # informar que a resposta é inválida e pedir para tentar novamente
else: # quando o contador inicial deixar de ser igual a 0:
    print('\nPrograma encerrado.') # imprimir na tela "programa encerrado"
```
