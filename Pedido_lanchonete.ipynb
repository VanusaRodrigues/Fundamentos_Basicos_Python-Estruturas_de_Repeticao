43. O cardápio de uma lanchonete é o seguinte: <br><br>
Especificação  --- Código ---  Preço <br>
    Cachorro Quente--- 100 ---    RS 1,20 <br>
    Bauru Simples---   101 ---    RS 1,30 <br>
    Bauru com ovo---   102 ---    RS 1,50 <br>
    Hambúrguer---      103 ---    RS 1,20 <br>
    Cheeseburguer---   104 ---    RS 1,30 <br>
    Refrigerante---    105 ---    RS 1,00 <br><br>
Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido. Considere que o cliente deve informar quando o pedido deve ser encerrado.

```python
lista_item = ['Cachorro Quente', 'Bauru Simples', 'Bauru com Ovo', 'Hambúrguer', 'Cheeseburger', 'Refrigerante'] # lista de itens
lista_codigo = [100, 101, 102, 103, 104, 105] # lista de código dos itens
lista_preco = [1.20, 1.30, 1.50, 1.20, 1.30, 1.00] # lista de preço unitário dos itens

lista_codigo_pedido = [] # lista que receberá os códigos dos itens informados no input
lista_item_pedido = [] # lista que receberá os nomes dos itens informados no input
lista_preco_unitario_pedido = [] # lista que receberá os preços unitários dos itens informados no input
lista_qtde_pedido = [] # lista que receberá as quantidades dos itens informados no input
lista_preco_total_pedido = [] # lista que receberá os preços * quantidades dos itens informados no input


while True: # enquanto verdade:
    codigo_item = int(input('\nInsira o código do item (pressione 0 para sair): ')) # pedir para informar o código do item
    if codigo_item == 0: # se o código informado for igual a 0
        print('Programa encerrado.') # informar encerramento na tela
        break # encerrar o programa
    elif codigo_item < 100 or codigo_item > 105:  # se o código do produto for menor que 100 ou maior que 105:
        print('\nCódigo inválido!') # informar que o código é inválido
    else: # se o código do produto for maior ou igual a 100 ou menor ou igual a 105:
        qtde_item = int(input('Informe a quantidade desejada: ')) # pedir a quantidade desejada
        for i, codigo in enumerate(lista_codigo): # para cada item na lista_codigo
            if codigo == codigo_item: # quando o código informado for igual ao código existente na lista
                preco_item = lista_preco[i] * qtde_item # calcular o preço a pagar pelo item (preço x quantidade)
                lista_codigo_pedido.append(lista_codigo[i]) # adicionar código do item informado na lista_codigo_pedido
                lista_item_pedido.append(lista_item[i]) # adicionar item na lista_item_pedido
                lista_preco_unitario_pedido.append(lista_preco[i]) # adicionar preço unitário na lista_preco_unitario_pedido
                lista_qtde_pedido.append(qtde_item) # adicionar quantidade informada na lista_qtde_pedido
                lista_preco_total_pedido.append(preco_item) # adicionar preço a pagar pelo item na lista_preco_total_pedido
                
                
for i in range(len(lista_codigo_pedido)): # para cada item da lista_codigo_pedido, imprimir as informações salvas sobre o item na tela
    print(f'\nCódigo: {lista_codigo_pedido[i]}')
    print(f'Item: {lista_item_pedido[i]}')
    print(f'Valor Unitário: R$ {lista_preco_unitario_pedido[i]:.2f}')
    print(f'Quantidade: {lista_qtde_pedido[i]}')
    print(f'Preço: R$ {lista_preco_total_pedido[i]:.2f}')

print(f'\nPreço total: R$ {sum(lista_preco_total_pedido):.2f}') # somar os preços de todos os itens e imprimir na tela
```
