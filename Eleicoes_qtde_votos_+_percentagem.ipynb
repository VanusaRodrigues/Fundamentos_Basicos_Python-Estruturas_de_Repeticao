44. Em uma eleição presidencial existem quatro candidatos. Os votos são informados por meio de código. Os códigos utilizados são:<br><br>
1 , 2, 3, 4  - Votos para os respectivos candidatos  <br>
(você deve montar a tabela ex: 1 - Jose/ 2- João/etc) <br>
5 - Voto Nulo <br>
6 - Voto em Branco <br><br>
Faça um programa que calcule e mostre: <br>
O total de votos para cada candidato; <br>
O total de votos nulos; <br>
O total de votos em branco; <br>
A percentagem de votos nulos sobre o total de votos; <br>
A percentagem de votos em branco sobre o total de votos. Para finalizar o conjunto de votos tem-se o valor zero.

```python
votos_1 = 0
votos_2 = 0
votos_3 = 0
votos_4 = 0
votos_5 = 0
votos_6 = 0
total_votos = 0

while True: # enquanto verdade:
    voto = int(input('''
ELEIÇÕES
Vote: 
1 - José
2 - João
3 - Maria
4 - Joana
5 - Voto Nulo
6 - Voto em Branco
Pressione 0 para sair: ''')) # pedir o voto ao usuário
    if voto == 0: # se voto for igual a 0:
        print('\nVotação encerrada.') # imprimir "votação encerrada"
        break # sair do loop while
    elif voto < 0 or voto > 6: # se voto informado não estiver entre 1 e 6:
        print('\nVoto inválido! Vote novamente.') # informar que é inválido e pedir para votar novamente
    else: # se voto informado estiver entre 1 e 6:
        total_votos += 1 # adicionar 1 voto ao contador total_votos
        if voto == 1: # se voto for para o candidato 1
            votos_1 += 1 # adicionar 1 voto ao contador votos_1
        elif voto == 2: # se voto for para o candidato 2
            votos_2 += 1 # adicionar 1 voto ao contador votos_2
        elif voto == 3: # se voto for para o candidato 3
            votos_3 += 1 # adicionar 1 voto ao contador votos_3
        elif voto == 4: # se voto for para o candidato 4
            votos_4 += 1 # adicionar 1 voto ao contador votos_4
        elif voto == 5: # se voto for nulo
            votos_5 += 1 # adicionar 1 voto ao contador votos_5
        elif voto == 6: # se voto for em branco
            votos_6 += 1 # adicionar 1 voto ao contador votos_6

# imprime total de votos, quantos votos cada candidato recebeu, quantos foram nulos e quantos foram em branco
print(f'''
Total de Votos: {total_votos}, sendo:
1 - José: {votos_1}
2 - João: {votos_2}
3 - Maria: {votos_3}
4 - Joana: {votos_4}
5 - Votos Nulos: {votos_5}
6 - Votos em Branco: {votos_6}''')

if total_votos > 0: # se ao menos uma pessoa votou (precisa fazer esta verificação para não dar erro de divisão por zero):
    # imprime percentagem de votos nulos e votos em branco sobre o total de votos
    print(f'''
\nA percentagem de votos nulos sobre o total de votos foi de {(votos_5 / total_votos):.2%}
\nA percentagem de votos em branco sobre o total de votos foi de {(votos_6 / total_votos):.2%}''') 
                  
```
