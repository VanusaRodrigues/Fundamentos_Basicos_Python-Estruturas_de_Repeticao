38. Um funcionário de uma empresa recebe aumento salarial anualmente: Sabe-se que: <br>
a) Esse funcionário foi contratado em 1995, com salário inicial de RS 1.000,00; <br>
b) Em 1996 recebeu aumento de 1,5% sobre seu salário inicial; <br>
c) A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior. Faça um programa que determine o salário atual desse funcionário. Após concluir isto, altere o programa permitindo que o usuário digite o salário inicial do funcionário.

```python
# salário inicial declarado na variável:

salario = 1000.00 # salário inicial (1995)
aumento = 0.015 # aumento inicial (1996)
ano_inicial = 1995
ano_atual = 2022

print(f'Ano: {ano_inicial} \nSalário inicial:R$ {salario:.2f}') # imprime ano inicial (1995) e salário inicial (R$ 1000.00)
salario += salario * aumento # calcula salário de 1996 com aumento de 1,5%
print(f'\nAno: {ano_inicial + 1} \nAumento: {aumento:.1%} \nSalário: R$ {salario:.2f}') # imprime ano 1996, aumento 1,5% e novo salário

while (ano_inicial + 1) < ano_atual: # enquanto o ano inicial não alcançar o ano atual:
    aumento += aumento # dobra o percentual do aumento
    salario += salario * aumento # aplica o novo aumento sobre o salário atual
    print(f'\nAno: {ano_inicial + 2} \nAumento: {aumento:.1%} \nSalário: R$ {salario:.2f}') # imprime ano atual, percentual e novo salário
    ano_inicial += 1 # acrescenta 1 no ano para reiniciar o loop e calcular o próximo ano

# * o cálculo, neste caso, está sendo efetuado sempre sobre o salário atual (não sobre os mil reais iniciais apenas).
# * assim, como o percentual dobra ano após ano, o aumento acaba sendo exorbitante.
```

```python
# salário inicial informado pelo usuário:

salario = float(input('Informe o salário inicial: R$ ')) # salário inicial (1995) informado pelo usuário
aumento = 0.015 # aumento inicial (1996)
ano_inicial = 1995
ano_atual = 2022

print(f'Ano: {ano_inicial} \nSalário inicial:R$ {salario:.2f}') # imprime ano inicial (1995) e salário inicial (R$ 1000.00)
salario += salario * aumento # calcula salário de 1996 com aumento de 1,5%
print(f'\nAno: {ano_inicial + 1} \nAumento: {aumento:.1%} \nSalário: R$ {salario:.2f}') # imprime ano 1996, aumento 1,5% e novo salário

while (ano_inicial + 1) < ano_atual: # enquanto o ano inicial não alcançar o ano atual:
    aumento += aumento # dobra o percentual do aumento
    salario += salario * aumento # aplica o novo aumento sobre o salário atual
    print(f'\nAno: {ano_inicial + 2} \nAumento: {aumento:.1%} \nSalário: R$ {salario:.2f}') # imprime ano atual, percentual e novo salário
    ano_inicial += 1 # acrescenta 1 no ano para reiniciar o loop e calcular o próximo ano

# * o cálculo, neste caso, está sendo efetuado sempre sobre o salário atual (não sobre os mil reais iniciais apenas).
# * assim, como o percentual dobra ano após ano, o aumento acaba sendo exorbitante.
```
