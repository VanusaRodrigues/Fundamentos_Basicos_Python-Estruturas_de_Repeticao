46. Em uma competição de salto em distância cada atleta tem direito a cinco saltos. No final da série de saltos de cada atleta, o melhor e o pior resultados são eliminados. O seu resultado fica sendo a média dos três valores restantes. Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe a média dos saltos conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média). Faça uso de uma lista para armazenar os saltos. Os saltos são informados na ordem da execução, portanto não são ordenados. O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo: <br><br>
Atleta: Rodrigo Curvêllo <br>
<br>
Primeiro Salto: 6.5 m <br>
Segundo Salto: 6.1 m <br>
Terceiro Salto: 6.2 m <br>
Quarto Salto: 5.4 m <br>
Quinto Salto: 5.3 m <br>
<br>
Melhor salto:  6.5 m <br>
Pior salto: 5.3 m <br>
Média dos demais saltos: 5.9 m <br>
<br>
Resultado final: <br>
Rodrigo Curvêllo: 5.9 m

```python
nome = input('Informe o nome do atleta (pressione a tecla Enter para sair): ') # pedir ao usuário para informar o nome do 1º atleta

lista_saltos = [] # lista que armazenará os saltos do atleta
lista_ordinais = ['Primeiro', 'Segundo', 'Terceiro', 'Quarto', 'Quinto'] # lista para utilizar ao imprimir na tela 'Primeiro salto', 'Segundo salto', etc

while nome != '': # enquanto o usuário não apertar enter no lugar de preencher o nome
    i = 1 # definir contador de saltos como 1
    while i <= 5: # enquanto 1 for menor ou igual a 5:
        salto = input(f'Informe a distância do {i}º salto (em metros): ') # pedir ao usuário para informar a distância do salto
        if salto == '': # se usuário apertar enter em vez de informar a distância
            print('Valor inválido! Tente novamente.') # informar que valor é inválido e pedir para tentar novamente
        else: # se usuário informar a distância corretamente
            lista_saltos.append(float(salto)) # adicionar a distância informada à lista_saltos
            i += 1 # acrescentar um ao contador de saltos
    else: # quando o contador i chegar a 6, parar de pedir a distância de saltos
        print(f'\nAtleta: {nome}\n') # imprimir na tela o nome do atleta
        for i, saltos in enumerate(lista_saltos): # para cada item da lista_saltos
            print(f'{lista_ordinais[i]} salto: {lista_saltos[i]} m') # imprimir na tela os 5 saltos
        
        print(f'\nMelhor salto: {max(lista_saltos)} m') # encontrar e imprimir na tela o melhor salto
        print(f'Pior salto: {min(lista_saltos)} m') # encontrar e imprimir na tela o pior salto
        
        lista_saltos.remove(max(lista_saltos)) # remover da lista_saltos o melhor salto
        lista_saltos.remove(min(lista_saltos)) # remover da lista_saltos o pior salto
        
        print(f'Média dos demais saltos: {sum(lista_saltos) / len(lista_saltos):.2f} m') # calcular e imprimir a média entre os 3 saltos restantes
        
        print(f'\nResultado final: \n{nome}: {sum(lista_saltos) / len(lista_saltos):.2f} m') # imprimir na tela resultado final (nome / média)
        
        lista_saltos.clear() # remover todos os itens da lista_saltos (para que não se somem valores do atleta anterior ao próximo a ser informado)
        
        # perguntar se deseja continuar e pedir nome do próximo atleta
        nome = input('\n\nDeseja continuar? \nInforme o nome do próximo atleta (pressione a tecla Enter para sair): ') 
else: # quando o usuário apertar enter no lugar de preencher o nome
    print('\nPrograma encerrado.') # imprimir na tela "Programa encerrado."
    
```
