47. Em uma competição de ginástica, cada atleta recebe votos de sete jurados. A melhor e a pior nota são eliminadas. A sua nota fica sendo a média dos votos restantes. Você deve fazer um programa que receba o nome do ginasta e as notas dos sete jurados alcançadas pelo atleta em sua apresentação e depois informe a sua média, conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média com as notas restantes). As notas não são informados ordenadas. Um exemplo de saída do programa deve ser conforme o exemplo abaixo: <br><br>
Atleta: Aparecido Parente <br><br>
Nota: 9.9 <br>
Nota: 7.5 <br>
Nota: 9.5 <br>
Nota: 8.5 <br>
Nota: 9.0 <br>
Nota: 8.5 <br>
Nota: 9.7 <br>
<br>
Resultado final: <br>
Atleta: Aparecido Parente <br>
Melhor nota: 9.9 <br>
Pior nota: 7.5 <br>
Média: 9,04 <br>

```python
nome = input('Informe o nome do atleta (pressione a tecla Enter para sair): ') # pedir ao usuário para informar o nome do 1º atleta

lista_notas = [] # lista que armazenará as notas do atleta

while nome != '': # enquanto o usuário não apertar enter no lugar de preencher o nome
    i = 1 # definir contador de notas como 1
    while i <= 7: # enquanto i for menor ou igual a 7:
        nota = input(f'Informe a {i}ª nota: ') # pedir ao usuário para informar a nota dada pelo jurado
        if nota == '': # se usuário apertar enter em vez de inserir a nota:
            print('Nota inválida! Tente novamente.') # informar que valor é inválido e pedir para tentar novamente
        elif nota != '': # se informar nota corretamente:
            nota = float(nota) # converter nota recebida como string em float
            if nota < 0 or nota > 10: # se usuário informar qualquer valor abaixo de 0 ou acima de 10:
                print('Nota inválida! Tente novamente.') # informar que valor é inválido e pedir para tentar novamente
            else: # se usuário informar a nota entre 0 e 10:
                lista_notas.append(nota) # adicionar a nota informada à lista_notas
                i += 1 # acrescentar 1 ao contador de notas
    else: # quando o contador i chegar a 8, parar de pedir notas
        print(f'\nAtleta: {nome}\n') # imprimir na tela o nome do atleta
        for i, notas in enumerate(lista_notas): # para cada item da lista_notas:
            print(f'Nota: {lista_notas[i]}') # imprimir na tela as notas dos 7 jurados
        
        print('\nResultado final:') # imprimir "Resultado final"
        print(f'Atleta: {nome}') # imprimir na tela o nome do atleta
        print(f'Melhor nota: {max(lista_notas)}') # encontrar e imprimir na tela a melhor nota
        print(f'Pior nota: {min(lista_notas)}') # encontrar e imprimir na tela a pior nota
        
        lista_notas.remove(max(lista_notas)) # remover da lista_notas a melhor nota
        lista_notas.remove(min(lista_notas)) # remover da lista_notas a pior nota
        
        print(f'Média: {sum(lista_notas) / len(lista_notas):.2f}') # calcular e imprimir a média entre as 5 notas restantes
        
        lista_notas.clear() # remover todos os itens da lista_notas (para que não se somem notas do atleta anterior ao próximo a ser informado)
        
        # perguntar se deseja continuar e pedir nome do próximo atleta
        nome = input('\n\nDeseja continuar? \nInforme o nome do próximo atleta (pressione a tecla Enter para sair): ') 
else: # quando o usuário apertar enter no lugar de preencher o nome
    print('\nPrograma encerrado.') # imprimir na tela "Programa encerrado."
    
```
